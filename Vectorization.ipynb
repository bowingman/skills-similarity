```python
from sklearn.feature_extraction.text import TfidfTransformer  
```

```python
tf = TfidfTransformer()
```

```python
X_tfidf = tf.fit_transform(res_array).toarray()
```

```python
np.fliplr(np.argsort(X_tfidf))
```

```python
X_tfidf
```

```python
from scipy.spatial.distance import cdist
```

```python
def cos_sim(A, B):
    similarities = 1 - cdist(A, B, metric='cosine')
    return similarities
```

```python
similarities = cos_sim(X_tfidf, X_tfidf)
```

```python
similarities
```

```python
similar = pd.DataFrame(similarities, columns=medium_res.columns)
```
