```python
res = similarity["programming"] + similarity["design"]
```

```python
def similar_skills(skills, similarity):
    res = similarity[skills[0]].copy()
    for skill in skills:
        res += similarity[skill]
    res -= similarity[skills[0]]
    return res
```

```python
res = similar_skills(["programming", "design", "science"], similarity)
```
